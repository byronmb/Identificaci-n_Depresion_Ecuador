```python
import pandas as pd
import numpy as np

tweets_depresivos_df = pd.DataFrame(pd.read_excel("Extraccion_Tweets/Dataset_depresivo_limpieza_manual.xlsx")) 

tweets_depresivos_df.head()
```

```python
#eliminamos los atributos que no son de utildad
tweets_depresivos_df.drop(['conversation_id', 'created_at', 'timezone', 'user_id', 'username', 'name', 'place', 'source', 'user_rt_id', 'user_rt', 'retweet_id', 'reply_to', 'retweet_date', 'translate', 'trans_src', 'trans_dest', 'language', 'mentions', 'urls', 'photos', 'replies_count', 'retweets_count', 'likes_count', 'hashtags', 'cashtags', 'link', 'retweet', 'quote_url', 'video', 'thumbnail', 'near', 'geo'], axis = 1, inplace = True)
tweets_depresivos_df.head()
tweets_depresivos_df = tweets_depresivos_df.drop_duplicates()
```

```python
tweets_depresivos_df = tweets_depresivos_df.sample(frac=1)
tweets_depresivos_df.to_excel('TweetsDepresivosValidacionPSIC.xlsx', index=False)
```
