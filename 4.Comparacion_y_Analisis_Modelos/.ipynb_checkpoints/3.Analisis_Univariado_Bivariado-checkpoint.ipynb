### Cantidad de publicaciones depresivas en base a los meses del año 2020 y 2021

```python
""" Separamos la columna "date" en campos que contengan solo el año y solo el mes para comparar 
    la cantidad de publicaciones """
tweets_2020_2021_df['año']= pd.DatetimeIndex(tweets_2020_2021_df['date']).year
tweets_2020_2021_df['mes']= pd.DatetimeIndex(tweets_2020_2021_df['date']).month
tweets_2020_2021_df.head()
```

```python
#Graficamos
def categorical_summarized(dataframe, x=None, hue=None, palette='Set1', verbose=True):
    sns.set(rc={'figure.figsize':(11,5)})
    g=sns.countplot(x=x, hue=hue, data=dataframe, palette=palette)
    g.set_xticklabels(['Enero','Febrero','Marzo','Abril','Mayo','Junio','Julio','Agosto','Septiembre','Octubre','Noviembre','Diciembre'],rotation = 20)
    g.set(xlabel='Mes',
       ylabel='Cantidad de Tweets',
       title='Publicaciones depresivas')
    plt.legend(title='Año')
    plt.rcParams["figure.facecolor"] = "w"
    plt.show()

# Feature Variable: Gender
c_palette = ['tab:blue', 'tab:orange']
#categorical_summarized(tweets_2020_2021_df, y = 'mes', palette=c_palette)
categorical_summarized(tweets_2020_2021_df, x = 'mes', hue='año', palette=c_palette)
```

###  Variación porcentual en el número de publicaciones depresivas desde el 2019 (prepandemia), hasta el 2020 y 2021 (en pandemia)

```python
#tweets predichos como depresivos por el modelo (año 2019)
tweets_2019_df = pd.DataFrame(pd.read_excel(path+"/4.Comparacion_y_Analisis_Modelos/Tweets_Depresivos_2019_Predichos.xlsx"))
# tweets depresivos (año 2020-2021)
tweets_20_21_df = pd.DataFrame(pd.read_excel(path+"/2.Preprocesamiento/Dataset_depresivo_Limpio.xlsx"))

#unimos ambos datasets para comparar
tweets_totales_df = pd.concat([tweets_2019_df,tweets_20_21_df])
#Eliminamos columnas inecesarias
tweets_totales_df.drop(['tweet_preprocesado', 'Sentimiento_Predicho', 'tweet_token'], axis = 1, inplace = True)
```

```python
#Obtenemos el año de la columna "date" para graficar la cantidad de publicaciones en base a cada año
tweets_totales_df['año']=pd.DatetimeIndex(tweets_totales_df['date']).year
#Guardamos las publicaciones de los 3 años(2019-2020-2021)
result_2019_2021=tweets_totales_df[(tweets_totales_df['año'] == 2019) | (tweets_totales_df['año'] == 2020) | (tweets_totales_df['año'] == 2021)]
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(8,5))
total = float(len(result_2019_2021))
ax = sns.countplot(x='año',data=result_2019_2021, palette="Set2")
plt.title('Cantidad de tweets depresivos prepandemia y durante la pandemia', fontsize=19)
ax.set(xlabel='Año',
   ylabel='Cantidad de Tweets')

for p in ax.patches: #porcentajes
   # percentage = '({:.1f}%)'.format(100 * p.get_height()/total)
    x = p.get_x() + p.get_width()
    y = p.get_height()+8
    #ax.annotate(percentage, (x, y),ha='right', fontsize=12)
for container in ax.containers: #valores
    ax.bar_label(container, fontsize=13)
plt.show()
```

### Comparación año predandemiay pandemia (2019-2020)

```python
result_2019_2020=tweets_totales_df[(tweets_totales_df['año'] == 2019) | (tweets_totales_df['año'] == 2020)]
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(8,5))
total = float(len(result_2019_2020))
ax = sns.countplot(x='año',data=result_2019_2020, palette="Set2")
plt.title('Variacion porcentual de Tweets', fontsize=19)
ax.set(xlabel='Año',
   ylabel='Cantidad de Tweets')

for p in ax.patches: #porcentajes
    percentage = '({:.1f}%)'.format(100 * p.get_height()/total)
    x = p.get_x() + p.get_width()
    y = p.get_height()+8
    ax.annotate(percentage, (x, y),ha='right', fontsize=12)
for container in ax.containers: #valores
    ax.bar_label(container, fontsize=13)
plt.show()
```

```python
#cantidad de tweets por año
tweets_por_año = result_2019_2020['año'].value_counts().sort_index()
tweets_por_año
```

```python
variacion_porcentual = ((tweets_por_año[2020] - tweets_por_año[2019]) / tweets_por_año[2019]) * 100
print(variacion_porcentual)
```

```python


# Calcula la cantidad de tweets por año
tweets_por_año = result_2019_2020['año'].value_counts().sort_index()

# Crea una lista que contenga los años
años = [str(año) for año in tweets_por_año.index]
print(años)
```

```python
################################
# Resetear el índice de la serie crecimiento_porcentual para asegurarnos de que sea numérico y esté en orden ascendente
crecimiento_porcentual = crecimiento_porcentual.reset_index(drop=True)

# Creación de una lista que contenga los años
años = [str(año) for año in tweets_por_año.index]

# Asegurarnos de que ambas listas tengan la misma longitud
min_length = min(len(años), len(crecimiento_porcentual))
años = años[:min_length]
crecimiento_porcentual = crecimiento_porcentual[:min_length]

# Calculamos el índice correspondiente al año 2020 en la lista de años
indice_2020 = años.index('2020')

# Creamos una nueva serie con los valores del crecimiento_porcentual más 100 en el eje vertical
line_data = crecimiento_porcentual + 100

# Creamos la figura y los ejes
fig, ax1 = plt.subplots(figsize=(10, 8))
ax2 = ax1.twinx()  # Segundo eje para mostrar la línea de crecimiento

# Graficamos las barras de cantidad de tweets por año
bars = ax1.bar(años, tweets_por_año, color='skyblue', label='Cantidad de Tweets')
ax1.set_ylabel('Cantidad de Tweets', color='dodgerblue')

# Graficamos la línea de crecimiento en la parte superior de las barras
line, = ax2.plot(años, line_data, marker='o', linestyle='-.', color='red', label='Crecimiento')
ax2.set_ylabel('Crecimiento', color='red')

# Agregamos etiquetas y título al gráfico
ax1.set_xlabel('Año')
ax1.set_title('Variación porcentual y cantidad de tweets por año')

# Mostramos el crecimiento porcentual en las etiquetas de la línea
ax2.text(0.4, 120, f'Variación: {crecimiento_porcentual.iloc[indice_2020]:.2f}%', ha='center')

# Calculamos el límite máximo del eje ax2 (crecimiento porcentual)
max_growth = max(crecimiento_porcentual)

# Establecemos los límites del eje y del ax2 para que empiece desde 0% hasta el 100% + el máximo crecimiento porcentual
ax2.set_ylim(0, max_growth + 106)

# Etiqueta en el punto de partida de la línea
start_value = crecimiento_porcentual.iloc[0] + 100
ax2.text(0, start_value , f'{start_value:.2f}%\n{tweets_por_año.iloc[0]} tweets',
         ha='right', va='bottom', color='maroon')

# Etiqueta en el punto final de la línea
end_x, end_y = len(años) - 1, 100 + crecimiento_porcentual.iloc[-1]
ax2.text(end_x, end_y, f'{crecimiento_porcentual.iloc[-1] + 100:.2f}%\n{tweets_por_año.iloc[1]} tweets',
         ha='left', va='bottom', color='maroon')

# Fusionar ambas leyendas en una sola y ubicarla en la parte izquierda
lines_labels = [ax.get_legend_handles_labels() for ax in fig.axes]
lines, labels = [sum(lol, []) for lol in zip(*lines_labels)]
ax1.legend(lines, labels, loc='upper left',facecolor='gainsboro')

plt.show()
#################################
```
